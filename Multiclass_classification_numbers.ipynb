```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
  plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train, y_train)
```

```python
model_score = model.score(X_test, y_test) * 100
print(f'The accuracy is: {model_score:.2f}%')
```

```python
plt.matshow(digits.images[67])
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

```python
# To check times where our model wasn't doing right
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python

import seaborn as sn
plt.figure(figsize = (10, 7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# From the above heatmap, 27 times the input was 0 and it predicted it to be 0.
# 1 time the actual was 6 and it predicted it to be 1
# 3 times the actual was 8 and it predicted it ot be 1
```
